```python
import os, getpass
from PIL import Image
import requests
from io import BytesIO
from langchain.llms import Replicate
```

```python
llm_generate_image = Replicate(
    model = "stability-ai/sdxl:1bfb924045802467cf8869d96b231a12e6aa994abfe37e337c63a4e49a8c6c41",
    model_kwargs = {"image_dimensions": "512x512"}
)

image_output = llm_generate_image("Iron Man flying next to the Eiffel Tower")
response = requests.get(image_output)
img = Image.open(BytesIO(response.content))

img
```
